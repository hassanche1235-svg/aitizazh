Task # 01:
Make a calculator in which you require the following inputs:
1. The first number
2. The second number
3. Operation to be performed
Also add a nested if for denominator == 0 in division

Task # 02:
Print the following stars pattern
*
**
***
****
Using for loop



```
def calculator(num1, num2, operator):
  if operator == '+' :
    print(num1 + num2)
  elif operator == '-' :
    print(num1 - num2)
  elif operator == '*' :
    print(num1 * num2)
  elif operator == '/' :
    if num2 == 0 :
      # nested if for division
      print('Error: Division by zero')
    else:
      print(num1 / num2)
  else:
    print('Error: Invalid operator')
calculator(10, 5, '/')
calculator(10, 5, '*')
calculator(10, 5, '+')
calculator(10, 5, '-')
calculator(10, 0, '/')
```
